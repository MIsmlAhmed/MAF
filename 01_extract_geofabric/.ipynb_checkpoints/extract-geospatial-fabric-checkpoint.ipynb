# Basic preparations

In this Notebook, the geospatial fabric draining to an outlet point (defined using lon and lat fields is extracted from the `MERIT-Basins` dataset.

Let's get started with our workflow and import necessary Python libraries:

```python
# load the extract functions from the `extract_geospfb_funs.py` file
from extract_geospfb_funs import *
```

# Define inputs

```python
# geofabric paths
## merit-basins provided in `rrg-mclark` - modify as needed
# merit_basins_root_path = '/project/rrg-mclark/data/geospatial-data/MERIT-Basins/MERIT_Hydro_v07_Basins_v01_bugfix1'
merit_basins_root_path = '/project/rpp-kshook/Climate_Forcing_Data/geospatial-data/MERIT-Basins/MERIT_Hydro_v07_Basins_v01_bugfix1'
# level 2 basin Pfafstetter code from HydroSHEDS/HydroBASINS 
# reference: https://www.reachhydro.org/home/params/merit-basins & https://data.hydrosheds.org/file/technical-documentation/HydroBASINS_TechDoc_v1c.pdf
# first digit represents continent code (7=North America, 8=Arctic region)
# second digit represents 9 large sub-units for 7 (1=Nelson, 2= St Lawrness great lakes, 8= West coast - BC)

# code = 71 # This is for the Bow at Banff and the North Sask River at Whirlpool (Nelson basin)
code = 82 # for smoky river

# outlet point for the watershed of interest
# create a Point object of long and lat
outlet_point = Point(-117.62308, 55.7146)


# output paths
output_path = '/home/mia725/scratch/calibration_workflow/Smoky_river/MERIT_geofabric/'
```

# Read MERIT Basins
**This part is static as it reads the entire MERIT geofabric (should only run once)**

```python
MERIT_catchments, MERIT_rivers, nca = read_MERIT_basins(merit_basins_root_path, code)
```

# Extract and plot the geofabric upstream of the specified outlet point

```python
extracted_catchments, extracted_rivers = extract_geofabric(MERIT_catchments, MERIT_rivers, outlet_point)
```

# Saving Extracted Geospatial Fabric upstream of outlet point

```python
save_extracted_geofabric(extracted_catchments, extracted_rivers, output_path)
```
